```python
import pandas as pd


train_path = "./data/titanic/train.csv"
test_path = "./data/titanic/test.csv"

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
```

```python
train_df = train_df[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
train_df["Sex"] = train_df["Sex"].astype("category").cat.codes
train_df["Embarked"] = train_df["Embarked"].astype("category").cat.codes
train_df.info()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

column_lst = list(train_df.columns)
column_lst.remove("Survived")

for column in column_lst:
    plt.figure()
    sns.regplot(train_df[[column, "Survived"]].groupby([column], as_index=False).mean().sort_values(column, ascending=False), x=column, y="Survived")
```
