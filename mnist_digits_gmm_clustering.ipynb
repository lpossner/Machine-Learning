```python
from sklearn.datasets import fetch_openml

X, y = fetch_openml("mnist_784", version=1, return_X_y=True, as_frame=False)
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(10, 10, figsize=(28, 28))
for idx, axi in enumerate(ax.flat):
    im = axi.imshow(X[idx].reshape(28, 28), cmap='binary')
```

```python
from tqdm import tqdm
import numpy as np
from sklearn.mixture import GaussianMixture

n_components = np.arange(50, 500, 10)
models = [GaussianMixture(n_components=n, covariance_type='full') for n in n_components]

aics = []
for model in tqdm(models):
    aics.append(model.fit(X[0:50]).aic(X))

# plt.plot(n_components, aics);
```
