```python
import numpy as np
import seaborn as sns
from scipy.stats import shapiro
```

```python
N = 1000
k = 0.5

X = np.linspace(-1, 1, N).reshape(N, -1)
Y = k * X + np.random.normal(loc=0, scale=0.5, size=(N, 1))

sns.scatterplot(x=X.squeeze(), y=Y.squeeze())
```

```python
# X --> Y

k_est = np.linalg.inv(X.T @ X) @ X.T @ Y

Y_est = k_est * X
Y_res = Y - Y_est

sns.scatterplot(x=X.squeeze(), y=Y_res.squeeze())
```

```python
# X --> Y

k_est = np.linalg.inv(Y.T @ Y) @ Y.T @ X

X_est = k_est * Y
X_res = X - X_est

sns.scatterplot(x=Y.squeeze(), y=X_res.squeeze())
```

```python
shapiro(Y_res)
```

```python
shapiro(X_res)
```
